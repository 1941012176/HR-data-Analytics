```python
ip = student.drop(['Extracurricular_Activities'],axis = 1) 
op = student.Performance_Index
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest, ytrain, ytest = train_test_split(ip,op,train_size=0.8)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc = StandardScaler()
mimx = MinMaxScaler()
```

```python
xtrain1 = sc.fit_transform(xtrain)
xtest1 = sc.fit_transform(xtest)
```

```python
print(xtrain1[0:5])
print(xtrain[0:5])
```

```python
xtrain2 = mimx.fit_transform(xtrain)
xtest2 = mimx.fit_transform(xtest)
```

```python
print('Standard',xtrain1[0:5])
print('Normal',xtrain[0:5])
print('mimx',xtrain2[0:5])
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xtrain1, ytrain)
```

```python
ypred = lr.predict(xtest1)
```

```python
ytest.shape, ypred.shape
```

```python
ypred
```

```python
ytest
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
mse = mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)
print('mean_Squared_Error=', mse, 'accuracy=', r2*100)
```

```python
sns.lmplot(x='Sample_Paper', y='Performance_Index',data=student)
plt.title('Simple Linear Regression Plot')
plt.show()
```

```python
sns.regplot(x = ytest, y= ypred,color='blue')
plt.show()
```

```python
r2Score = r2_score(ytest,ypred)
print('Accuracy = ',round(r2Score,4)*100,'%')
mae_score = mean_absolute_error(ytest,ypred)
print('Mean Absolute Error = ',round(mae_score,4)*100,'%')
rmse_score = np.sqrt(mean_squared_error(ytest,ypred))
print('Root Mean Squared Error = ',round(rmse_score,4)*100,'%')
```

```python
plt.scatter(ytest, ypred, color='blue')
plt.plot([min(ytest), max(ytest)], [min(ytest), max(ytest)], color='red', linewidth=2)
plt.xlabel('Actual Performance Index')
plt.ylabel('Predicted Performance Index')
plt.title('Actual vs Predicted Performance Index')
plt.show()
```
